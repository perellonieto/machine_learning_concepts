```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
import datasets
```

```python
for n_features in range(10, 500, 30):
    all_s = []
    all_y = []
    for i in range(10):
        nif = int(n_features/2)
        ncpc = int(n_features/5)
        x, y = make_classification(n_samples=10000, n_features=n_features,
                                   n_informative=nif, n_clusters_per_class=ncpc)
        ss = StandardScaler()
        x = ss.fit_transform(x)
        clf = LinearSVC(dual=False)
        clf.fit(x, y)
        all_s.append(clf.decision_function(x))
        all_y.append(y)
    plt.figure()
    plt.title('#features={}'.format(n_features))
    all_s = np.concatenate(all_s)
    all_y = np.concatenate(all_y)
    plt.hist(all_s[all_y==0], alpha=0.4, bins=50, color='blue')
    plt.hist(all_s[all_y==1], alpha=0.4, bins=50, color='red')
```
