```python
import matplotlib.pyplot as plt
import numpy as np

# create some randomly ddistributed data:
data = np.random.randn(10000)

# sort the data:
data_sorted = np.sort(data)

# calculate the proportional values of samples
p = 1. * np.arange(len(data)) / (len(data) - 1)

# plot the sorted data:
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(p, data_sorted)
ax1.set_xlabel('$p$')
ax1.set_ylabel('$x$')

ax2 = fig.add_subplot(122)
ax2.plot(data_sorted, p)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$p$')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create some randomly ddistributed data:
data = np.concatenate([np.random.randn(1000),
                       np.random.triangular(8, 10, 11, 1000)])

# sort the data:
data_sorted = np.sort(data)

# calculate the proportional values of samples
p = 1. * np.arange(len(data)) / (len(data) - 1)

# plot the sorted data:
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(p, data_sorted)
ax1.set_xlabel('$p$')
ax1.set_ylabel('$x$')

ax2 = fig.add_subplot(122)
ax2.plot(data_sorted, p)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$p$')
```

```python
fig = plt.figure(figsize=(13, 5))
ax = fig.add_subplot(1, 2, 1)
ax.set_title('Original data')
ax.hist(data, bins=100)

ax = fig.add_subplot(1, 2, 2)
p = np.random.ranf(1000)
indices = (p*len(data)).astype(int)
ax.set_title('Samples from inverse')
ax.hist(data_sorted[indices], bins=100);
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.random.randn(10000) # generate samples from normal distribution (discrete data)
x = np.sort(x)
norm_cdf = norm.cdf(x) # calculate the cdf - also discrete

# plot the cdf
plt.plot(x, norm_cdf)
plt.show()
```
